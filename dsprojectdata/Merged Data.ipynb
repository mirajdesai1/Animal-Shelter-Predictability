```python
# separate the date and time of intake
def separate_date(time):
    full = str(time)
    date = full[0:10]

    return date

def separate_time(time):
    full = str(time)
    time = full[11:]

    return time
```

```python
austin_outcomes["Outcome Date"] = austin_outcomes["DateTime"].apply(separate_date)
austin_outcomes["Outcome Time"] = austin_outcomes["DateTime"].apply(separate_time)
austin_outcomes.drop(["DateTime","MonthYear"],axis=1, inplace=True)
austin_outcomes
```

```python
austin_intakes["Intake Date"] = austin_intakes["DateTime"].apply(separate_date)
austin_intakes["Intake Time"] = austin_intakes["DateTime"].apply(separate_time)
austin_intakes.drop(["DateTime","MonthYear"],axis=1, inplace=True)
austin_intakes
```

```python
merged_austin = pd.merge(austin_intakes, austin_outcomes, left_on=["Animal ID","Color","Animal Type","Breed","Name"],
                        right_on=["Animal ID","Color","Animal Type","Breed","Name"])
merged_austin.drop(["Found Location"],axis=1, inplace=True)
merged_austin
```

```python
#making the outcomes standardixe
def change_outcome(outcome):
    if (outcome == "Return to Owner"):
        return "Returned to owner"
    elif (outcome == 'Euthanasia'):
        return "Euthanized"
def change_intake(outcome):
    if (outcome == "Dispos req"):
        return "Euthanasia Request"
def change_condition(outcome):
    if (outcome == "Healthy"):
        return "Normal"
    elif(outcome == "Deceased"):
        return "Dead"
```

```python
def change_time(time):
    full = str(time)
    hour = int(time[0:2])
    rest = time[2:]
    day = ""
    if (hour > 12):
        hour = str(0) + str(hour - 12)
        day = " PM"
    else:
        day = " AM"
    return str(hour) + str(rest) + str(day)
```

```python
dallas["Intake Time"] = dallas["Intake Time"].apply(change_time)
dallas["Outcome Time"] = dallas["Outcome Time"].apply(change_time)
```

```python
months = {1:31,2:28,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}
monthsleap = {1:31,2:29,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}
```

```python
def return_days(intake, outcome): 
    if (type(intake) == str and type(outcome) == str): 
    
        intake_month = int(intake[:2])
        intake_day = int(intake[3:5])
        intake_year = int(intake[6:])
    
        outcome_month = int(outcome[:2])
        outcome_day = int(outcome[3:5])
        outcome_year = int(outcome[6:])
    
        if (outcome_year == intake_year):
            if (outcome_month == intake_month):
                return int(outcome_day - intake_day + 1) #same month
            else: 
                if (outcome_year % 4 == 0):
                    return int((monthsleap[intake_month] - intake_day + 1) + getAllDays(intake_month, int(outcome_month), monthsleap) + outcome_day)
                else:
                    return int((months[intake_month] - intake_day + 1) + getAllDays(intake_month, int(outcome_month), months) + outcome_day)
            
        else: #year isn't the same
            if (outcome_year - intake_year == 1):
                daysAfterIntake = return_days(intake, "12/31/" + str(intake_year))
                daysToOutcome = return_days("01/01/" + str(outcome_year),outcome) 
                return int(daysAfterIntake + daysToOutcome)
            else: 
                daysAfterIntake = return_days(intake, "12/31/" + str(intake_year))
                daysToOutcome = return_days("01/01/" + str(outcome_year),outcome) 
                middleYears = getYears(intake_year+1,outcome_year)
                return int(daysAfterIntake + daysToOutcome +  middleYears)
```

```python
def getAllDays(in_m,out_m,cal):
    between = out_m - in_m
    days = 0
    for i in range(1, between):
        days += cal[in_m + i] 
    return days
```

```python
def getYears(in_y,out_y):
    days = 0
    for i in range(in_y,out_y):
        if(i % 4 == 0):
            days += 366
        else:
            days +=365
    return days
            
```

```python
merged["Days In"] = merged.apply(lambda x: return_days(x['Intake Date'], x['Outcome Date']), axis=1)
```
