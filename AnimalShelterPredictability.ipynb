```python
# This method returns the # of days that the animal spent in the shelter
# This is calculated from the Intake Date and Outcome Dates. It includes the the day in, and the day out
def return_days(intake, outcome): 
    if (type(intake) == str and type(outcome) == str): 
    
        intake_month = int(intake[:2])
        intake_day = int(intake[3:5])
        intake_year = int(intake[6:])
    
        outcome_month = int(outcome[:2])
        outcome_day = int(outcome[3:5])
        outcome_year = int(outcome[6:])
    
        if (outcome_year == intake_year):
            if (outcome_month == intake_month):
                return int(outcome_day - intake_day + 1) #same month
            else: 
                if (outcome_year % 4 == 0):
                    return int((monthsleap[intake_month] - intake_day + 1) + getAllDays(intake_month, int(outcome_month), monthsleap) + outcome_day)
                else:
                    return int((months[intake_month] - intake_day + 1) + getAllDays(intake_month, int(outcome_month), months) + outcome_day)
            
        else: #year isn't the same
            if (outcome_year - intake_year == 1):
                daysAfterIntake = return_days(intake, "12/31/" + str(intake_year))
                daysToOutcome = return_days("01/01/" + str(outcome_year),outcome) 
                return int(daysAfterIntake + daysToOutcome)
            else: 
                daysAfterIntake = return_days(intake, "12/31/" + str(intake_year))
                daysToOutcome = return_days("01/01/" + str(outcome_year),outcome) 
                middleYears = getYears(intake_year+1,outcome_year)
                return int(daysAfterIntake + daysToOutcome +  middleYears)
```

```python
# This method calculates the number of days between months
def getAllDays(in_m,out_m,cal):
    between = out_m - in_m
    days = 0
    for i in range(1, between):
        days += cal[in_m + i] 
    return days
```

```python
# This method calculates the number of days between years
def getYears(in_y,out_y):
    days = 0
    for i in range(in_y,out_y):
        if(i % 4 == 0):
            days += 366
        else:
            days +=365
    return days
```

```python
from scipy import stats

def independent_t_samples(df, iv1, iv2, iv_name, dv, series):
    """ conducts a independent t samples test"""
    iv1 = df[df[iv_name] == iv1][dv]
    iv2 = df[df[iv_name] == iv2][dv]
    
    descriptives = df.groupby(iv_name).agg(["count","mean","std","sem"])
    results = stats.ttest_ind(iv1, iv2) #signficant
    t_stat = results[0]
    p_value = results[1]
    format(p_value,".10f")
    df = len(iv1) + len(iv2) - 2
    levene_results = stats.levene(iv1,iv2)
    shapiro_results = stats.shapiro(series)
    

    print("levene results: ", end = "")
    print(levene_results)
    print("shapiro results: ", end = "")
    print(shapiro_results)
    print ("t(%d)=%.2f,p=%.3f" %(df,t_stat,p_value))
    
    print("Descriptives: \n")
    return descriptives
```

```python
from scipy import stats
from statsmodels.stats.multicomp import MultiComparison
from statsmodels.stats.multicomp import pairwise_tukeyhsd

def oneway_ANOVA(data, IV, DV):
    anova_title = "-----------------------\nONE-WAY ANOVA RESULTS\n-----------------------\n\n"
    ftest_title = "F-test\n-------\n"
    assump_title = "Assumption Checks\n-------------------\n\n"
    EOV_title = "Assumption of Equality of Variances:\n"
    AON_title = "Assumption of Normality:\n"
    PH_title = "\nPost-hoc Tests\n----------------\n\n"
    # f-stat, df calculations
    f_stat, p_value = getFStat(data,IV,DV)
    f_stat, p_value = format(f_stat, ".2f"), format(p_value, ".4f")
    df1, df2 = getDf(data, IV, DV)
    f_stat_info = ("F(%s,%s) = %s, p = %s" % (df1,df2,f_stat,p_value)) + "\n\n"
    # levene calculations
    levene, levene_cond = check_EOV(data,IV,DV)
    levene_info = (("\tLeveneResult(statistic=%s, pvalue=0.%s)\n\t" + levene_cond + "\n\n") % (levene[0], levene[1]))
    # shapiro info
    shapiro_info = check_AON(data,IV,DV)
   
    print(anova_title + ftest_title + f_stat_info 
          + assump_title + EOV_title + levene_info + AON_title + shapiro_info, end ="")
    
    mc = MultiComparison(data[DV],data[IV])
    tukey_result = mc.tukeyhsd()
    print(PH_title, tukey_result)
    
```

```python
def getFStat(data,IV,DV):
    iv_agg = data[IV]
    iv_names = iv_agg.unique()
    iv_specific = []
    for i in range(0,len(iv_names)):
        dv = data[iv_agg == iv_names[i]][DV]
        iv_specific.append(dv)
    return stats.f_oneway(*(data[data[IV] == name][DV] for name in iv_names))
```

```python
def getDf(data,IV,DV):
    """Returns the two degrees of freedom for an between-subjects
    experiment with 3+ independent levels and 1 depedent variable"""
    iv_agg = data[IV]
    iv_names = iv_agg.unique()
    df1 = len(iv_names) - 1
    df2 = 0;
    
    for i in range(0,len(iv_names)):
        df2 += (len(data[iv_agg == iv_names[i]][DV]) - 1)
        
    return (df1,df2)
```

```python
def check_AON(data,IV,DV):
    """Returns the assumption of normality, and whether it is met for
    an between-subjects experiment with 3+ independent levels and 1 depedent variable"""
    iv_agg = data[IV]
    iv_names = iv_agg.unique()
    iv_specific = []
    message = ""
    for i in range(0,len(iv_names)):
        dv = data[iv_agg == iv_names[i]][DV]
        iv_specific.append(dv)
        shapiro = stats.shapiro(iv_specific[i])
        message = (message + "\t" + iv_names[i] + " : (%s,%s)" % (shapiro[0],shapiro[1]))
        if (shapiro[1] > .05):
            message = message + "\n\t\tAssumption is met. p > .05\n\n"
        else:
            message = message + "\n\t\tAssumption is violated. p < .05\n\n"
    return message
```

```python
def check_EOV(data,IV,DV):
    iv_agg = data[IV]
    iv_names = iv_agg.unique()
    iv_specific = []
    for i in range(0,len(iv_names)):
        dv = data[iv_agg == iv_names[i]][DV]
        iv_specific.append(dv)
    results = stats.levene(*(data[data[IV] == name][DV] for name in iv_names))
    if (results[1] > .05): #assumption is met
        return (results,"Assumption is met. p > .05")
    else:
        return (results,"Assumption is violated. p < .05")
```
